```python
import pandas as pd

# Load your dataset
df = pd.read_csv("/content/drive/MyDrive/CausalDigitalTwin/driver_behavior_route_anomaly_dataset_with_derived_features.csv")  # change path if needed

# Dictionary to store unique values info
unique_summary = {}

for col in df.columns:
    unique_vals = df[col].dropna().unique()
    unique_summary[col] = {
        "num_unique": len(unique_vals),
        "unique_values": unique_vals
    }

# Convert summary to DataFrame for readability
summary_df = pd.DataFrame.from_dict(unique_summary, orient="index")

print(summary_df)
```

```python
import pandas as pd

pd.set_option("display.max_colwidth", None)
pd.set_option("display.max_rows", None)

df = pd.read_csv(
    "/content/drive/MyDrive/CausalDigitalTwin/driver_behavior_route_anomaly_dataset_with_derived_features.csv"
)

summary_rows = []

for col in df.columns:
    unique_vals = df[col].dropna().unique().tolist()
    summary_rows.append({
        "column": col,
        "num_unique": len(unique_vals),
        "unique_values": unique_vals
    })

summary_df = pd.DataFrame(summary_rows)
print(summary_df)
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv(
    "/content/drive/MyDrive/CausalDigitalTwin/driver_behavior_route_anomaly_dataset_with_derived_features.csv"
)

for col in df.columns:
    unique_vals = df[col].dropna().unique()
    print("=" * 80)
    print(f"Column: {col}")
    print(f"Number of unique values: {len(unique_vals)}")
    print("Unique values:")
    for val in unique_vals:
        print(val)
```

```python
rows = []

for col in df.columns:
    for val in df[col].dropna().unique():
        rows.append({
            "column": col,
            "unique_value": val
        })

unique_df = pd.DataFrame(rows)
unique_df.to_csv("/content/drive/MyDrive/CausalDigitalTwin/Results/unique_values_per_column.csv", index=False)

print("Saved to unique_values_per_column.csv")
```
